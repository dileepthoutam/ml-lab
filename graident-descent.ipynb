```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('Housing.csv')
df.head()
```

```python
x = df.iloc[:, 2:3]
y = df.iloc[:, 1]
```

```python
learning_rate = 0.05
iterations = 2000
```

```python
np.random.seed(44)
x = np.array(sorted(list(range(5))*20)) + np.random.normal(size=100, scale=0.5)
y = np.array(sorted(list(range(5))*20)) + np.random.normal(size=100, scale=0.25)
n = x.shape[0]
m = 1
c = 1
m_gradients, c_gradients, costs = [], [], []
for i in range(iterations):
    m_gradient = (-2) * (np.sum(x * (y - (m*x+c)))) / n
    c_gradient = (-2) * (np.sum(y * (y - (m*x + c)))) / n
    m -= learning_rate * m_gradient
    c -= learning_rate * c_gradient
    m_gradients.append(m)
    c_gradients.append(c)
    cost = np.sum((y-(m*x+c))**2) / n
    costs.append(cost)

def predict(x):
    return m * x + c
plt.scatter(x, y, color='black')
plt.plot(x, predict(x))
plt.show()
```
