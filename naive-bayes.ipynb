```python
import pandas as pd
import numpy as np
```

```python
data = pd.read_csv('weather.nominal.csv')
data.head()
```

```python
from sklearn.preprocessing import LabelEncoder
lc = LabelEncoder()

data['outlook'] = lc.fit_transform(data['outlook'])
data['temperature'] = lc.fit_transform(data['temperature'])
data['humidity'] = lc.fit_transform(data['humidity'])
data['windy'] = lc.fit_transform(data['windy'])
data['play'] = lc.fit_transform(data['play'])
data.head()
```

```python
features = list(zip(data['outlook'], data['temperature'], data['humidity'], data['windy']))
label = data['play']
```

```python
from sklearn.naive_bayes import GaussianNB
model = GaussianNB()

model.fit(features, label)
```

```python
predicted = model.predict([[0, 1, 0, 0]])
print(predicted)
```
