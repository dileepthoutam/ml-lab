```python
import pandas as pd
```

```python
df = pd.read_csv('diabetes.csv')
```

```python
x = df.drop("Outcome",axis=1)
y = df["Outcome"]
```

```python
from keras.models import Sequential
from keras.layers import Dense
model = Sequential()

model.add(Dense(12,input_dim=8,activation="relu"))
model.add(Dense(12,activation="relu"))
model.add(Dense(12,activation="relu"))
model.add(Dense(1,activation="sigmoid"))
```

```python
model.compile(loss="binary_crossentropy",optimizer="adam",metrics=["accuracy"])
```

```python
model.fit(x,y,epochs=150,batch_size=10)
```

```python
loss, accuracy = model.evaluate(x, y)
print("Model loss: %.2f"% (loss*100))
print("Model accuracy: %.2f"% (accuracy*100))
```

```python
model.evaluate(x,y)
```

```python
predictions = model.predict(x)
print([round(x[0]) for x in predictions])
```

```python
for i in range(len(predictions)):
  x = predictions[i]
  predictions[i] = round(x[0])
print("Actual, Predicted, Result")
for i in range(len(y)):
  if predictions[i]==y[i]:
    ans = True
  else:
    ans = False
  print(y[i],predictions[i],ans)
```

```python
from sklearn.metrics import confusion_matrix
arr = confusion_matrix(y,predictions,labels=[1,0])
print("Confusion matrix = \n",arr)
tn,fp,fn,tp = arr.ravel()
```

```python
from sklearn.metrics import precision_score,recall_score,f1_score
precision = precision_score(y,predictions)
recall = recall_score(y,predictions)
f1 = f1_score(y,predictions,average=None)
print("Precision = ",precision)
print("Recall = ",recall)
print("F1 score = ",f1)
```
