1. Read csv file
2. Perform label encoding column wise (use fit_transform)
3. Combine features using zip (tuple returns)
4. import KNeighborsClassifier
5. fit the model with (features, label)
5. predict the value

```python
import numpy as np
import pandas as pd
```

```python
data = pd.read_csv('weather.nominal.csv')
```

```python
data.head()
```

```python
from sklearn import preprocessing
le = preprocessing.LabelEncoder()
```

```python
data['outlook'] = le.fit_transform(data['outlook'])
data['temperature'] = le.fit_transform(data['temperature'])
data['humidity'] = le.fit_transform(data['humidity'])
data['windy'] = le.fit_transform(data['windy'])
data['play'] = le.fit_transform(data['play'])
```

```python
data.head()
```

```python
features = list(zip(data['outlook'], data['temperature'], data['humidity'], data['windy']))
label = data['play']
```

```python
from sklearn.neighbors import KNeighborsClassifier
model = KNeighborsClassifier(n_neighbors=5)
```

```python
model.fit(features, label)
```

```python
pred = model.predict([[2, 2, 1, 0]])
pred
```
