**Design a Function to implement concept of Gradient Descent Algorithm for Linear
Regression**

**AIM:** To implement concept of gradient descent algorithm for linear regression

**DESCRIPTION:** 
Gradient Descent is an optimization algorithm used for minimizing the cost function in various machine learning algorithms. It is basically used for updating the parameters of the learning model.

It is a function that measures the performance of the model for any given data. After making
a hypothesis, we calulate the cost function. Main goal is to reduce the cost function, we
modify the parameters by using the Gradient Descent Algorithm over the given data.
Gradient descent is an iterative optimization algorithm to and the minimum of a function.
Here the function is cost function.

Gradient of a function


Derivatives of a function shows how much a value changes when you change its argument.
Derivatives are important for optimization because the saddle derivatives might indicate a
maxmimum, minimum or saddle point.

**Steps**
1. Compute the gradient, the first order derivative of the function at that point.
2. Make a step in the direction opposite to the gradient, opposite direction of slope
increase from the current point by alpha times the gradient at that point.


```python
import numpy as np

def gradient_descent(x,y):
  m_curr = c_curr = 0
  iterations = 50
  n = len(x)
  learning_rate = 0.08
  for i in range(iterations):
    y_predicted = m_curr*x + c_curr
    cost = (1/n)* sum([val**2 for val in (y-y_predicted)])
    md = -(2/n)*sum(x*(y-y_predicted))
    cd = -(2/n)*sum(y-y_predicted)

    m_curr = m_curr - learning_rate * md
    c_curr = c_curr - learning_rate * cd
    print("m {},c {},cost {},i {}".format(m_curr, c_curr, cost, i ))
x = np.array([1,2,3,4,5])
y = np.array([5,7,9,11,13])
gradient_descent(x,y)
```

**Gradient descent algorithm using Housing.csv dataset and linear regression**

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
learning_rate = 0.05
iterations = 2000
```

```python
df = pd.read_csv("Housing.csv")
df.head()
```

Assume lotsize as x and price as y. Now for finding the best fit line that describes the best
relationship between price and lotsize.
Let the best fit line be y = mx + c

Goal is to find the values of m and c using gradient descent algorithm.


```python
X = df.iloc[::, 2]
X.head()
```

```python
y = df.iloc[::, 1]
y.head()
```

```python
np.random.seed(44)
X = np.array(sorted(list(range(5))*20)) + np.random.normal(size=100, scale=0.5)
y = np.array(sorted(list(range(5))*20)) + np.random.normal(size=100, scale=0.25)
```

```python
n = X.shape[0]
m = 1
c = 1
m_gradients, c_gradients, costs = [], [], []
for i in range(iterations):
 m_gradient = (-2)*(np.sum(X * (y - (m*X + c))))/n
 c_gradient = (-2)*(np.sum(y - (m*X + c)))/n
 m -= learning_rate * m_gradient
 c -= learning_rate * c_gradient
 m_gradients.append(m)
 c_gradients.append(c)
 cost = np.sum((y - (m*X + c))**2)/n
 costs.append(cost)
 print ("m {}, c {}, cost {} iteration {}".format(m,c,cost, i+1))
```

```python
def predict(x):
 return m * x + c
```
